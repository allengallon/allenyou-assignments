```python
#total expenses and revenue per state, this goes into Flourish
eada22_23.groupby('STATENAME')[['GRND_TOTAL_EXPENSE']].sum().to_csv('state_exp.csv')
eada22_23.groupby('STATENAME')[['GRND_TOTAL_REVENUE']].sum().to_csv('state_rev.csv')
```

```python
#addresses to input into geocodio
addresses = eada22_23[['addr1_txt', 'city_txt', 'state_cd', 'zip_text']]

addresses.to_csv('addresses.csv')

#reading the geocodio output
lat_long = pd.read_csv('latlong.csv')

eada22_23['Latitude'] = lat_long['Geocodio Latitude']
eada22_23['Longitude'] = lat_long['Geocodio Longitude']

#mapping the highest expenditure school per state, goes into Flourish
max_idx = eada22_23.groupby('STATENAME')['GRND_TOTAL_EXPENSE'].idxmax()
eada22_23.loc[max_idx][['STATENAME','institution_name', 'GRND_TOTAL_EXPENSE', 'Latitude', 'Longitude']].to_csv('high_state_exp.csv')

#mapping the highest expenditure school per state, goes into Flourish
max2_idx = eada22_23.groupby('STATENAME')['GRND_TOTAL_REVENUE'].idxmax()
eada22_23.loc[max2_idx][['STATENAME','institution_name', 'GRND_TOTAL_REVENUE', 'Latitude', 'Longitude']].to_csv('high_state_rev.csv')
```

```python
#sports expenditures totals among all schools, in Flourish I removed redundant rows that were accounted for in combined rows
#(Cross Country, Track Indoors, Track Outdoors, Swimming, Diving)

per_sport = schools_data.groupby('Sports')[['EXPENSE_MENALL', 'EXPENSE_WOMENALL', 'TOTAL_EXPENSE_ALL']].sum().sort_values(
    by = 'TOTAL_EXPENSE_ALL', ascending = False)

per_sport.to_csv('per_sport.csv')
```

```python
#sports revenue totals among all schools, in Flourish I removed redundant rows that were accounted for in combined rows
#(Cross Country, Track Indoors, Track Outdoors, Swimming, Diving)

per_sport_rev = schools_data.groupby('Sports')[['REVENUE_MENALL', 'REVENUE_WOMENALL', 'TOTAL_REVENUE_ALL']].sum().sort_values(
    by = 'TOTAL_REVENUE_ALL', ascending = False)

per_sport_rev.to_csv('per_sport_rev.csv')
```

```python
#import previous year's data
schools_data_2122 = pd.read_excel('EADA_2021-2022/Schools.xlsx')
```

```python
#group 21-22 data by sport
per_sport22 = schools_data_2122.groupby('Sports')[['EXPENSE_MENALL', 'EXPENSE_WOMENALL', 'TOTAL_EXPENSE_ALL']].sum()

#merge with 22-23 data set
merged = pd.merge(per_sport, per_sport22, on = 'Sports')

#combine volleyball and beach volleyball
merged.loc['Volleyball & Beach Volleyball'] = merged.loc['Volleyball'] + merged.loc['Beach Volleyball']

merged = merged.drop(['Beach Volleyball', 'Volleyball'])

#percent change calculation
pct_change = pd.DataFrame()

pct_change['Men'] = (merged.EXPENSE_MENALL_x - merged.EXPENSE_MENALL_y)/merged.EXPENSE_MENALL_y
pct_change['Women'] = (merged.EXPENSE_WOMENALL_x - merged.EXPENSE_WOMENALL_y)/merged.EXPENSE_WOMENALL_y
pct_change['Total'] = (merged.TOTAL_EXPENSE_ALL_x - merged.TOTAL_EXPENSE_ALL_y)/merged.TOTAL_EXPENSE_ALL_y

#melt into long-form data for Flourish
pct_change.melt(ignore_index = False).to_csv('pct_change.csv')
```
